```python
data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'BSc Computer Science_Male': [42, 36, 38, 33, 34, 37, 35, 41, 29],
    'BSc Computer Science_Female': [11, 14, 13, 11, 15, 10, 11, 12, 16],
    'BSc Microprocessor Instrumentation and Technology_Male': [24, 24, 20, 25, 23, 25, 24, 25, 21],
    'BSc Microprocessor Instrumentation and Technology_Female': [3, 8, 9, 8, 7, 5, 6, 5, 9],
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

# To reshape the DataFrame in a way that separates the gender from the program, 

```python

# Restructure the DataFrame using pd.melt()
df1 = pd.melt(df, id_vars=['Year'], var_name='Course_Gender', value_name='Enrollment')

# Separate 'Course_Gender' column into 'Course' and 'Gender' columns
df1[['Course', 'Gender']] = df1['Course_Gender'].str.split('_', expand=True)

# Drop the original 'Course_Gender' column
df1 = df1.drop('Course_Gender', axis=1)

# Display the head of the restructured DataFrame
df1.head(10)
```

```python
# Save the reshaped DataFrame to a CSV file
df1.to_csv('enrollment_data.csv', index=False)
# Display the head of the restructured DataFrame
df1.head(20)
```

# To compare total enrollments between the Microprocessor Instrumentation course and the Computer Science course using a pie chart, 

```python

# Calculate total enrollments for each course
total_enrollments = df1.groupby('Course')['Enrollment'].sum()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_enrollments, labels=total_enrollments.index, autopct='%1.1f%%', startangle=140)
plt.title('Comparison of Total Enrollments by Course using a pie-chart')
plt.show()

# Calculate total enrollments for Computer Science and Microprocessor Instrumentation
total_comp_science_enrollment = df1.loc[df1['Course'] == 'BSc Computer Science', 'Enrollment'].sum()
total_microprocessor_enrollment = df1.loc[df1['Course'] == 'BSc Microprocessor Instrumentation and Technology', 'Enrollment'].sum()

# Calculate the sum total for all enrollments
sum_total_enrollment = df1['Enrollment'].sum()

# Display the results
print("Total Computer Science Enrollment:", total_comp_science_enrollment)
print("Total Microprocessor Instrumentation Enrollment:", total_microprocessor_enrollment)
print("Sum Total Enrollment for All Courses:", sum_total_enrollment)
```

# To create a pie chart showing the comparison of the popularity of courses based on the number of graduates, you can follow these steps:

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df_reshaped is your DataFrame with the reshaped graduation data

# Clean the 'Gender' column by removing leading and trailing whitespaces
df_reshaped['Gender'] = df_reshaped['Gender'].str.strip()

# Aggregate the data by program and calculate the total number of graduates
total_graduates_by_program = df_reshaped.groupby('Program')['Number_of_Graduates'].sum().reset_index()

# Sort the programs by the total number of graduates in descending order
sorted_programs = total_graduates_by_program.sort_values(by='Number_of_Graduates', ascending=False)

# Plot a pie chart
plt.figure(figsize=(10, 10))
plt.pie(sorted_programs['Number_of_Graduates'], labels=sorted_programs['Program'], autopct='%1.1f%%', startangle=90)
plt.title('Popularity of Courses based on Graduates')
plt.show()
```

# To generate a circle graph (also known as a donut chart) showcasing the comparison between male and female graduates over the three-year period

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df_reshaped is your DataFrame with the reshaped graduation data

# Clean the 'Gender' column by removing leading and trailing whitespaces
df_reshaped['Gender'] = df_reshaped['Gender'].str.strip()

# Aggregate the data by gender and calculate the total number of graduates
total_graduates_by_gender = df_reshaped.groupby('Gender')['Number_of_Graduates'].sum().reset_index()

# Plot a donut chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(total_graduates_by_gender['Number_of_Graduates'], labels=total_graduates_by_gender['Gender'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
ax.set_title('A Circle graph showing comparison of Male and Female Graduates (3-Year Period)')
plt.show()
```

```python
# code for total of most graduates for an individual course
import matplotlib.pyplot as plt
import seaborn as sns

# Clean the 'Gender' column by removing leading and trailing whitespaces
df_reshaped['Gender'] = df_reshaped['Gender'].str.strip()

# Filter the DataFrame for the 'Bachelors in Computer Science' program
bachelors_cs_data = df_reshaped[df_reshaped['Program'] == "Bachelor's of science in computer science"]

# Find the year with the highest total graduates
max_year = bachelors_cs_data.groupby('Year')['Number_of_Graduates'].sum().idxmax()

# Filter the DataFrame for the specific year with the highest total graduates
max_year_data = bachelors_cs_data[bachelors_cs_data['Year'] == max_year]

# Plot a bar graph for the 'Bachelors in Computer Science' program for the year with the highest total graduates
plt.figure(figsize=(8, 6))
sns.barplot(x='Year', y='Number_of_Graduates', hue='Gender', data=max_year_data)
plt.title(f"Total Graduates in Bachelor's of science in computer science ({max_year}) by Gender")
plt.xlabel('Year')
plt.ylabel('Number of Graduates')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Print the total number of graduates for each gender in the year with the highest total graduates
total_males = max_year_data[max_year_data['Gender'] == 'Male']['Number_of_Graduates'].sum()
total_females = max_year_data[max_year_data['Gender'] == 'Female']['Number_of_Graduates'].sum()

print(f"Total Graduates in Bachelor's of science in computer science ({max_year}):")
print(f'Total Males: {total_males}')
print(f'Total Females: {total_females}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean the 'Gender' column by removing leading and trailing whitespaces
df_reshaped['Gender'] = df_reshaped['Gender'].str.strip()

# Filter the DataFrame for the 'Bachelors in Computer Science' program
bachelors_cs_data = df_reshaped[df_reshaped['Program'] == 'Bachelors in Computer Science']

# Check if the DataFrame is not empty
if not bachelors_cs_data.empty:
    # Find the year with the highest total graduates
    max_year = bachelors_cs_data.groupby('Year')['Number_of_Graduates'].sum().idxmax()

    # Filter the DataFrame for the specific year with the highest total graduates
    max_year_data = bachelors_cs_data[bachelors_cs_data['Year'] == max_year]

    # Plot a bar graph for the 'Bachelors in Computer Science' program for the year with the highest total graduates
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Year', y='Number_of_Graduates', hue='Gender', data=max_year_data)
    plt.title(f'Total Graduates in Bachelors Computer Science ({max_year}) by Gender')
    plt.xlabel('Year')
    plt.ylabel('Number of Graduates')
    plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

    # Print the total number of graduates for each gender in the year with the highest total graduates
    total_males = max_year_data[max_year_data['Gender'] == 'Male']['Number_of_Graduates'].sum()
    total_females = max_year_data[max_year_data['Gender'] == 'Female']['Number_of_Graduates'].sum()

    print(f'Total Graduates in Bachelors Computer Science ({max_year}):')
    print(f'Total Males: {total_males}')
    print(f'Total Females: {total_females}')
else:
    print('No data available for the Bachelors in Computer Science program.')
```
